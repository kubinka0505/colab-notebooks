```python
#@title ## Closest value to given BPM that can be used for *Dancing Waveform Experiment*
#@markdown Float values are estimated.
BPM = 130.000 #@param {type:"number"}

#-=-=-=-#

def frange(start = 0, stop = None, step = 1):
	"""Range with float support."""
	while start < stop:
		yield round(start, 3)
		start += step

def closest_bpm(target_bpm, step = 0.001, max_bpm = None):
	if max_bpm is None:
		max_bpm = target_bpm

	closest_value = None
	min_diff = float("inf")

	for value in frange(1, max_bpm, step):
		rounded_value = round(value, 3)
		bpm = rounded_value

		if len(str(bpm / 60).split(".")[-1]) <= 2:
			diff = abs(bpm - target_bpm)

			if diff < min_diff:
				min_diff = diff
				closest_value = bpm

	return closest_value

#-=-=-=-#

print(closest_bpm(BPM), "BPM")
```

```python
#@title ## Time stretch or pitch shift required to convert `Base` BPM to `Target` BPM.

Base = 135 #@param {type: "number"}
Target = 120 #@param {type: "number"}
Output_Advanced_Equation = False #@param {type: "boolean"}

#@markdown ---

#@markdown > ‚ÑπÔ∏è Works with sample rates too.

from math import *

#-=-=-=-#

Result_Time = round(100 * (Base / Target), 3)
Result_Pitch = ceil((1200 / log(2)) * log(Target / Base))

if Output_Advanced_Equation:
	Result_Time_Display = f"‚åä100 √ó ({Base} √∑ {Target})‚åâ = {Result_Time}"
	Result_Pitch_Display = f"‚åà(1200 √∑ log(2)) √ó log({Target} √∑ {Base})‚åâ = {Result_Pitch}"
else:
	Result_Time_Display = Result_Time
	Result_Pitch_Display = Result_Pitch

#-=-=-=-#

print("Time stretch:\t" + f"{Result_Time_Display}%")
print("Pitch shift:\t" + f"{Result_Pitch_Display} cents")
```

```python
#@title ## Resample by cents.
#@markdown Uses following formula: $\left\lfloor SampleRate \times 2^{\left(\frac{Resample\_by}{1200}\right)} \right\rceil$

#@markdown ---

SampleRate = 44100 #@param {type: "number"}
Resample_by = -100 #@param {type: "number"}
Output_Advanced_Equation = False #@param {type: "boolean"}

#@markdown ---

#@markdown | Parameter | Description |
#@markdown |:-|:-|
#@markdown | $Samplerate$ | Base audio samplerate |
#@markdown | $Resample\_by$ | Cents |

#-=-=-=-#

Sign = "+ " if Resample_by > 0 else "- "
Resample_by_Display = str(abs(Resample_by))

Frequency = round(SampleRate * (2 ** (Resample_by / 1200)))

#-=-=-=-#

if Output_Advanced_Equation:
	retval = f"‚åä{SampleRate} Hz √ó 2 ^ ({Sign.strip()}{Resample_by_Display} cents √∑ 1200)‚åâ = {Frequency} Hz"
else:
	retval = f"{SampleRate} Hz {Sign}{Resample_by_Display} cents = {Frequency} Hz"

print(retval)
```
